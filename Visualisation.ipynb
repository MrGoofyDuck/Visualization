# Bar chart 

```python
import matplotlib.pyplot as plt
years=range( 2013, 2019)
apples=[0.5, 0.3, 0.9, 0.1, 0.8, 0.3]
pear=[0.8, 0.9, 0.6, 0.2, 0.7, 0.6]

plt.bar(years, apples)

plt.xlabel('Year')
plt.ylabel('Yield(ton per hectare)')

plt.title('Crop Yields on Kanto')
```

```python
plt.bar(years, apples)
plt.bar(years, pear, bottom=apples)
```

# Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt

y=np.array([45, 20, 55, 10])

alabels=['Apple', 'Pear', 'Peach', 'Cherry']

nexplode=[0.2, 0, 0, 0]

plt.pie(y, labels=alabels, explode=nexplode)
```

# 3D

```python
import numpy as np
import matplotlib.pyplot as plt 
ax=plt.axes(projection="3d")

x_data=np.arange(-5, 5, 0.1)
y_data=np.arange(-5, 5, 0.1)

x, y= np.meshgrid(x_data, y_data)
z=np.sin(x)*y

ax.plot_surface(x, y, z)
plt.show()
```
